```python
class Person :
    def __init__(self , name):
        self.name = name 

    def show_name(self) :
        print(f"Name : {self.name}")
        
        
```

```python
class Student(Person):
    def __init__(self, name , grade):
        super().__init__(name)
        self.grade = grade 
    def status(self):
        if self.grade >= 10 :
            return "pass"
        else : 
            return "fail"    
    def show_info(self):
           print(f"Name : {self.name} , Grade: {self.grade} , Status : {self.status}"  ) 
```

```python
def save_students(student):
    with open ("students.txt" , "a") as file:
        file.write(f"{student.name} , {student.grade}\n")
```

```python
def load_students():
    students = []
    try:
        with open("students.txt", "r") as file:
            for line in file:
                name, grade = line.strip().split(",")
                students.append(Student(name, int(grade)))
    except FileNotFoundError:
        print("File not found!")
    return students
```

```python
s1 = Student("Ali", 15)
s2 = Student("Reza", 9)

save_students(s1)
save_students(s2)

students = load_students()

for s in students:
    s.show_info()
```
