```python
students = {}

while True:
    print("\n1. Add student")
    print("2. Show all students")
    print("3. Search student")
    print("4. Class average")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Student name: ")
        n = int(input("Number of scores: "))
        scores = []

        for i in range(n):
            s = float(input("Score: "))
            scores.append(s)

        students[name] = scores
        print("Student added")

    elif choice == "2":
        for name, scores in students.items():
            print(name, ":", scores)

    elif choice == "3":
        name = input("Enter name: ")
        if name in students:
            scores = students[name]
            avg = sum(scores) / len(scores)
            print("Scores:", scores)
            print("Average:", avg)
        else:
            print("Student not found")

    elif choice == "4":
        total = 0
        count = 0

        for scores in students.values():
            total += sum(scores)
            count += len(scores)

        if count > 0:
            print("Class average:", total / count)
        else:
            print("No data")

    elif choice == "5":
        break

    else:
        print("Invalid choice")


```
