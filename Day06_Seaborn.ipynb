```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./iris.csv")
```

```python
df
```

```python
sns.histplot(data=df ,x="petal_length" , hue="species" , bins=11 , alpha = 0.5)
plt.xlabel("petal")
plt.ylabel("distribution")
plt.title("hist")
plt.show()
```

```python
sns.barplot(data=df , x="species" , y="petal_length",hue="species" , estimator="mean" , palette=["red" , "green" , "blue"])
```
